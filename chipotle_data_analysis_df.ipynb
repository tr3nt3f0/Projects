```python
import pandas as pd #as keyword is used to define an alias

url = "https://raw.githubusercontent.com/braincomputingsantosh/Data/master/chipotle.tsv.txt"
chipoDF = pd.read_csv(url, sep='\t') #t=tab 

chipoDF
```

# Chipotle Dataframes

```python
#delete the duplicates in item_name and quantity
chipoDF_filtered = chipoDF.drop_duplicates(['item_name', 'quantity'])
chipoDF_filtered
```

```python
#clean the items_price column and trnasform it in a float
prices = [float(value[1:-1]) for value in chipoDF.item_price]
prices
```

```python
chipoDF.item_price = prices
```

```python
#select only the products with quantity equals to 1
chipoDF_one_product = chipoDF_filtered[chipoDF_filtered.quantity == 1]
chipoDF_one_product
```

```python
chipoDF_one_product[chipoDF_one_product['item_price']<10].item_name.nunique()
```

```python
#what is the price of each time?
# orient data frame with only 2 columns, item_name and item_price

chipoDF_filtered = chipoDF.drop_duplicates(['item_name', 'quantity'])
#select only the products with quantity equals 1
chipoDF_one_product = chipoDF_filtered[chipoDF_filtered.quantity == 1]
#Select only the item_name an item_price columns
price_per_item = chipoDF_one_product[['item_name', 'item_price']]
#sort the values from the most to less expensive
price_per_item.sort_values(by=["item_name", "item_price"], ascending=[True, False])#.head(10)
```
